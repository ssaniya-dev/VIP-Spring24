```python
from datetime import datetime
from datetime import timedelta
import asyncio
import aiohttp

async def fetch(session, url, params=None):
    async with session.get(url, params=params) as response:
        return await response.json()
```

```python
from dateutil.relativedelta import relativedelta
async def collect_data():
    tasks = []
    async with aiohttp.ClientSession() as session:
        for keyword in keywords:
            cur_datetime = datetime(year=2004, month=1, day=1)
            while cur_datetime.year < 2016:
                params = {
                    "engine": "google_trends",
                    "q": keyword,
                    "date": f"{cur_datetime.strftime('%Y-%m-%d')} {(cur_datetime + relativedelta(months=1) - relativedelta(days=1)).strftime('%Y-%m-%d')}",
                    "tz": "420",
                    "data_type": "TIMESERIES"
                }
                tasks.append(fetch(session, "https://serpapi.com/search?engine=google_trends", params=params))
                cur_datetime += relativedelta(months=1) 
                print(tasks)
                return await asyncio.gather(*tasks)
        print(tasks)
        return await asyncio.gather(*tasks)
                
```

```python
from collections import defaultdict

result = defaultdict(list)
for keyword in keywords:
    for year in range(2004, 2016):
        cur_dt = f"{year}-01-01"
        end_dt = f"{year}-12-31"
        params = {
            "engine": "google_trends",
            "q": keyword,
            "date": f"{cur_dt} {end_dt}",
            "tz": "420",
            "data_type": "TIMESERIES",
            #"geo": "US"
        }
        try:
            res = requests.get("https://serpapi.com/search?engine=google_trends", params=params).json()
            result[keyword].append(res)
            print(res)
        except Exception as e:
            print(f"Failed to get data for {keyword} in {year}")
            print(e)
        
```

```python
trend = TrendReq()

trend.build_payload(kw_list=['Inflation'], timeframe='2004-01-01 2005-01-01')
trend.interest_over_time()
```

```python
from pytrends.exceptions import TooManyRequestsError

result = defaultdict(lambda: defaultdict(list))
sleep_time = 60
import time
for keyword in keywords:
    for year in range(2004, 2016):
        cur_dt = f"{year}-01-01"
        end_dt = f"{year}-12-31"
        trend.build_payload(kw_list=[keyword], timeframe=f'{cur_dt} {end_dt}')
        try:
            result[keyword][year] = trend.interest_over_time()
            print(result[keyword][year])
            sleep_time = 60
        except TooManyRequestsError as e:
            print(f"Too many requests for {keyword} in {year}")
            time.sleep(sleep_time)
            sleep_time += 60
            year -= 1
        except Exception as e:
            print(f"Failed to get data for {keyword} in {year}")
            print(e)
            raise e
```
